```python
#Imports
import pandas as pd
```

```python
#Import the csv file

df = pd.read_csv('Zip_Code_Median_Sale_Prices_Seasonally_Adjusted.csv')
df.head()
```

```python
#Create a new dataframe of the housing fields needed for this project. Since there are only two columns not needed
#(Region Id and Size), delete these columns
df2 = df.drop(['RegionID','SizeRank'],axis=1)
df2.head()
```

```python
#Rename the columns: only one column name (RegionName) needs to be renamed to "ZipCode".
df2.columns = ['ZipCode' if x=='RegionName' else x for x in df2.columns]

df2.head(2)
```

```python
#Write the dataframe out to a csv file
out_csv = 'project1_housing_data.csv'

df2.to_csv(out_csv)
```
