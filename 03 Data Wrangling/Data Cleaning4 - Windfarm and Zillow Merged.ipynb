```python
#Imports
import pandas as pd
```

```python
#Load the cleaned zillow csv file to a dataframe
%cd C:\Users\Ken\Documents\KenP\Applications-DataScience\SpringboardCourseWork\CapstoneProject1Repository\09 CapstoneProject1Data\ZillowData

df_zi = pd.read_csv('project1_housing_data.csv')
#df_zi = pd.read_csv('junk_zi2.csv')

df_zi = df_zi.sort_values(by=['ZipCode'])

df_zi.head(1)
```

```python
#Load the cleaned & expanded windfarm csv file to a dataframe
%cd C:\Users\Ken\Documents\KenP\Applications-DataScience\SpringboardCourseWork\CapstoneProject1Repository\09 CapstoneProject1Data\WindfarmData

df_wf = pd.read_csv('project1_windfarm_data_expanded.csv')
#df_wf = pd.read_csv('junk_wf2.csv')

df_wf = df_wf.sort_values(by=['zipcode'])

df_wf.head(1)
```

```python
#Change directory to MergedData

%cd C:\Users\Ken\Documents\KenP\Applications-DataScience\SpringboardCourseWork\CapstoneProject1Downloads\MergedData
```

```python
#Merge the windfarm and zillow dataframes on zip code, as inner join, and write to a csv file.

df_me = pd.merge(left=df_wf, right=df_zi, on=None, left_on='zipcode', right_on='ZipCode')
df_me2 = df_me.drop(df_me.columns[1],axis=1)
df_me2 = df_me2.drop(df_me2.columns[0],axis=1)
```

```python
df_me2.head(1)
```

```python
out_csv = 'merged_wi_and_zi.csv'
df_me2.to_csv(out_csv)
```
