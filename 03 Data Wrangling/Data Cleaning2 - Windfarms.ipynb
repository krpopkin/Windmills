```python
#Imports
import pandas as pd
import numpy as np
```

```python
#Import the csv file

df = pd.read_csv('uswtdb_v1_2_20181001.csv')
df.head()
```

```python
#Create a new dataframe of the fields needed for this project
df2 = df[['t_state','t_county','t_fips','p_name','p_year','p_tnum','p_cap','xlong','ylat']]
df2.head()
```

```python
#Rename the columns
column_labels_list = ['state', 'county', 'fips', 'project_name', 'project_op_year',
                      'windmills_count', 'capacity', 'longitude', 'latitude']

df2.columns = column_labels_list
df2.head(2)
```

```python
#Group the rows by "project_name" (windfarm level)

df3 = df2
df3 = df3.reset_index()
df3 = df3.set_index('project_name')

df3 = df2.groupby(['state','county','fips','project_name'])
```

```python
#Write the dataframe out to a csv file
out_csv = 'project1_windfarm_data_grouped_by_project_name.csv'
df3.sum().reset_index().to_csv(out_csv)
```

```python
#Evaluate the data for cleanliness
df2.info()
```

```python
#Evaluate data for cleanliness (continued)

#Identify cells populated with the word 'missing'
state_bad_data = len(df2[df2.state == 'missing'])
print('state bad data: ' + str(state_bad_data))

county_bad_data = len(df2[df2.county == 'missing'])
print('county bad data: ' + str(county_bad_data))

fips_bad_data = len(df2[df2.fips == 0])
print('fips id bad data: ' + str(fips_bad_data))

project_name_bad_data = len(df2[df2.project_name == 'missing'])
print('project name bad data: ' + str(project_name_bad_data))

project_op_year_bad_data = len(df2[df2.project_op_year == 0])
print('project op year id bad data: ' + str(project_op_year_bad_data))

windmills_count_bad_data = len(df2[df2.windmills_count == 0])
print('windmills count bad data: ' + str(windmills_count_bad_data))

capacity_bad_data = len(df2[df2.capacity == 0])
print('capacity bad data: ' + str(capacity_bad_data))

longitude_bad_data = len(df2[df2.longitude == 0])
print('longitude bad data: ' + str(longitude_bad_data))

latitude_bad_data = len(df2[df2.latitude == 0])
print('latitude bad data: ' + str(latitude_bad_data))
```
